```python
from bs4 import BeautifulSoup
import requests
```

```python
wiki = requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
wiki
```

```python
soup = BeautifulSoup(wiki.content)
soup
```

```python
titles = []

for i in soup.find_all(['h1', 'h2','h3','h4','h5','h6']):
    titles.append(i.text)
    
print(titles)
```
