### 4) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) 
from https://presidentofindia.nic.in/former-presidents.htm

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://presidentofindia.nic.in/former-presidents.htm")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
name =[]

for i in soup.find_all("h3"):
    name.append(i.text)
    
print(name)
```

```python
term =[]

for i in soup.find_all("div",class_="presidentListing"):
    term.append(i.text.split(":")[1].replace("\n","").replace("http",""))
    
print(term)
```

```python
len(name),len(term)
```

```python
president_data= {"Name":name,"Term":term}
president_df = pd.DataFrame(president_data)
president_df
```
