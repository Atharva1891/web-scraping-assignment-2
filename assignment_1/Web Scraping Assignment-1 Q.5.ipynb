5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
a) Top 10 ODI teams in menâ€™s cricket along with the records for matches, points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
```

```python
soup
```

 a) Top 10 ODI teams in menâ€™s cricket along with the records for matches, points and rating.

```python
team = []

for i in soup.find_all("span",class_="u-hide-phablet"):
    team.append(i.text)
    
print(team)
```

```python
end_match = soup.find("td",class_="rankings-block__banner--matches").text
print(end_match)
end_point = soup.find("td",class_="rankings-block__banner--points").text
print(end_point)
end_rating = soup.find("td",class_="rankings-block__banner--rating u-text-right").text
end_rating=end_rating.replace(" ","").replace("\n","")
print(end_rating)
```

```python
m2 = []

for i in soup.find_all("td",class_="table-body__cell u-center-text"):
    m2.append(i.text)
    
m2
```

```python
matches = m2[0::2]
matches
```

```python
points = m2[1::2]
points
```

```python
matches.insert(0,end_match)
matches
```

```python
points.insert(0,end_point)
points
```

```python
rating = []

for i in soup.find_all("td",class_="table-body__cell u-text-right rating"):
    rating.append(i.text)
    
rating
```

```python
rating.insert(0,end_rating)
```

```python
data = {"Team":team[:10],"Matches":matches[0:10],"Point":points[:10],"Rating":rating[:10]}
data
```

```python
# a) Top 10 ODI teams in menâ€™s cricket along with the records for matches, points and rating.
odi_team = pd.DataFrame(data)
odi_team
```
